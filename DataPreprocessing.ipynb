```python
dir_food101 = "/Users/aniketindulkar/Downloads/food-101/Dataset-V2/images"
list_of_food = listdir(dir_food101)
food_dataset = '/Users/aniketindulkar/Downloads/food-101/Dataset-V2/dataset'
dataset_size = 500
train_size = 400
test_size = 100
```

```python
list_of_food = sorted(list_of_food)
if '.DS_Store' in list_of_food:
    list_of_food.remove('.DS_Store')
```

```python
ingredients = []
all_ingredients =[]
food_ingredient_dict = {}
classes = 0
food_index = 0
with open("/Users/aniketindulkar/Downloads/food-101/Dataset-V2/Ingredients_v2.txt", 'r') as file:
    for line in file:
        line = line.rstrip('\n')
        ingredient_list = []
        for word in line.split(","):
            word = word.strip().lower()
            ingredient_list.append(word)
            all_ingredients.append(word)
        ingredients.append(ingredient_list)
        food_ingredient_dict[list_of_food[food_index]]= ingredient_list
        food_index= food_index+1

classes = len(set(all_ingredients))
classes_l = list(set(all_ingredients))
```

```python
cooking_type = []
all_cooking_type =[]
food_cooking_type_dict = {}
classes = 0
food_index = 0
with open("/Users/aniketindulkar/Downloads/food-101/Dataset-V2/CookingType.txt", 'r') as file:
    for line in file:
        line = line.strip().lower()
        all_cooking_type.append(line)
        cooking_type.append(line)
        food_cooking_type_dict[list_of_food[food_index]]= line
        food_index= food_index+1

cooking_classes = len(set(all_cooking_type))
```

```python
def makeDir(directory):
    if os.path.exists(directory):
        shutil.rmtree(directory)
    os.makedirs(directory)
```

```python
total_count = 0
train_count = 0
test_count =0
item_count =0;
train_csv_data ={}
test_csv_data = {}
makeDir(food_dataset)
```

```python
# to get ingredients as List in one Column
for food in list_of_food:
    if isdir(dir_food101+"/"+food): # if you want all the data
        for image_path in listdir(dir_food101+"/"+food):
            if not image_path.startswith('.'):
                image = cv2.imread(os.path.join(dir_food101+"/"+food,image_path))
                height, width = image.shape[0],image.shape[1]
                image = image [int(0.2*height):int((height - (height*0.2))),int(0.2*width):int((width - (0.2*width)))]
                image = cv2.resize(image,(128,128),interpolation = cv2.INTER_AREA)                
                cv2.imwrite(food_dataset+"/"+food+str(total_count)+".jpg",image)
                if train_count < train_size:  
                    train_csv_data[food+str(total_count)+".jpg"]= ingredients[item_count]
                    train_count= train_count+1
                elif test_count < test_size:
                    test_csv_data[food+str(total_count)+".jpg"]= ingredients[item_count]
                    test_count= test_count+1
                else:
                    break
                total_count=total_count+1
        test_count = 0
        train_count = 0
        item_count= item_count+1
```

```python
# To get all the ingredient as column

for food in list_of_food:
    if isdir(dir_food101+"/"+food): # if you want all the data
        for image_path in listdir(dir_food101+"/"+food):
            if not image_path.startswith('.'):
                image = cv2.imread(os.path.join(dir_food101+"/"+food,image_path))
                height, width = image.shape[0],image.shape[1]
                image = image [int(0.2*height):int((height - (height*0.2))),int(0.2*width):int((width - (0.2*width)))]
                image = cv2.resize(image,(128,128),interpolation = cv2.INTER_AREA)                
                cv2.imwrite(food_dataset+"/"+food+str(total_count)+".jpg",image)
                label_list = []
                for label in classes_l:
                    if label in ingredients[item_count]:
                        label_list.append(1)
                    else:
                        label_list.append(0)
                if train_count < train_size:
                    train_csv_data[food+str(total_count)+".jpg"]= label_list
                    train_count= train_count+1
                elif test_count < test_size:
                    test_csv_data[food+str(total_count)+".jpg"]= label_list
                    test_count= test_count+1
                else:
                    break
                total_count=total_count+1
        test_count = 0
        train_count = 0
        item_count= item_count+1
```
